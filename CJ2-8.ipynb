import numpy, matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

1. Create a numpy array x ranging from [0,1] in 100 steps
2. Define a funtioon that returns sin(x) and cos(x)

```python
x = np.linspace(0,1,100)

def ang(n):
    return np.sin(n), np.cos(n)
```

Create a multipanel plot
plotting sin(x) vs x in the left panel and
in the left panel cos(x) vs x in the right panel
Label respective panels and save the figure as a pdf

```python
y_sin, y_cos = ang(x)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(6,6))

ax1.plot(x,y_sin,'r-')
ax1.set_title("sin(x) vs. x")
ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")
ax1.grid(True)

ax2.plot(x,y_cos,'g-')
ax2.set_title("cos(x) vs. x")
ax2.set_xlabel("x")
ax2.set_ylabel("cos(x)")
ax2.grid(True)




plt.savefig("CJ2-8.pdf")
plt.show()
```
