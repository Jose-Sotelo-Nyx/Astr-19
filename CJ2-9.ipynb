Make a histogram with 100 random numbers ditributed uniformly between [0,1].

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

make numpy give 100 random numbers that are distributed uniformly

```python
n = 1000 

x = np.random.randn(n) 



width = 0.1
hist_min = np.floor(min(x)) 
hist_max = np.ceil(max(x)) + width 

bins =np.arange(hist_min,hist_max,width)


plt.hist(x,bins=121,alpha=0.5,edgecolor='g')
plt.xlabel('x')
plt.ylabel('Count')
print(len(bins)-1)
plt.show()
```
