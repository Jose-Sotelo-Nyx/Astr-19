Make a histogram with 100 random numbers ditributed uniformly between [0,1].

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

make numpy give 100 random numbers that are distributed uniformly

```python
n = 1000 

x = np.random.randn(n) 



width = 0.1
hist_min = np.floor(min(x)) 
hist_max = np.ceil(max(x)) + width 

# contains the border of the bins
bins =np.linspace(hist_min,hist_max,101)

# plot the histogram
plt.hist(x,bins=bins,alpha=0.5,edgecolor='g')
plt.xlabel('x')
plt.ylabel('Count')
plt.savefig("CJ2-9.pdf")
plt.show()
```
